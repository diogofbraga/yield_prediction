# Gather GNN Results

### Import Libraries

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_rows = None
pd.set_option('display.float_format', lambda x: '%.4f' % x)
```

### Collect Data

```python
xl_additive1 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/additive/ranking_test1/results.xlsx")
xl_additive2 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/additive/ranking_test2/results.xlsx")
xl_additive3 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/additive/ranking_test3/results.xlsx")
xl_additive4 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/additive/ranking_test4/results.xlsx")

xl_aryl1 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/aryl_halide/ranking_test1/results.xlsx")
xl_aryl2 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/aryl_halide/ranking_test2/results.xlsx")
xl_aryl3 = pd.ExcelFile("results/graph_descriptors/WL_gnn/out_of_sample/aryl_halide/ranking_test3/results.xlsx")
```

```python
xl_additive1.sheet_names
```

```python
df_additive1 = xl_additive1.parse("Sheet1")
df_additive2 = xl_additive2.parse("Sheet1")
df_additive3 = xl_additive3.parse("Sheet1")
df_additive4 = xl_additive4.parse("Sheet1")

df_aryl1 = xl_aryl1.parse("Sheet1")
df_aryl2 = xl_aryl2.parse("Sheet1")
df_aryl3 = xl_aryl3.parse("Sheet1")
```

```python
df_additive1
```

#### Check if they follow the same structure

```python
df1 = df_additive1[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]
df2 = df_additive2[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]
df3 = df_additive3[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]
df4 = df_additive4[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]

if df1.equals(df2) and df2.equals(df3) and df3.equals(df4):
    print(True)
else:
    print(False)
```

```python
df1 = df_aryl1[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]
df2 = df_aryl2[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]
df3 = df_aryl3[['num_layers', 'learning_rate', 'graph_readout', 'molecules_combination']]

if df1.equals(df2) and df2.equals(df3):
    print(True)
else:
    print(False)
```

```python
df_additive1.rename(columns={'rmse_train': 'rmse_train1'}, inplace=True)
df_additive2.rename(columns={'rmse_train': 'rmse_train2'}, inplace=True)
df_additive3.rename(columns={'rmse_train': 'rmse_train3'}, inplace=True)
df_additive4.rename(columns={'rmse_train': 'rmse_train4'}, inplace=True)

df_additive1.rename(columns={'r2_train': 'r2_train1'}, inplace=True)
df_additive2.rename(columns={'r2_train': 'r2_train2'}, inplace=True)
df_additive3.rename(columns={'r2_train': 'r2_train3'}, inplace=True)
df_additive4.rename(columns={'r2_train': 'r2_train4'}, inplace=True)

df_additive1.rename(columns={'rmse_test': 'rmse_test1'}, inplace=True)
df_additive2.rename(columns={'rmse_test': 'rmse_test2'}, inplace=True)
df_additive3.rename(columns={'rmse_test': 'rmse_test3'}, inplace=True)
df_additive4.rename(columns={'rmse_test': 'rmse_test4'}, inplace=True)

df_additive1.rename(columns={'r2_test': 'r2_test1'}, inplace=True)
df_additive2.rename(columns={'r2_test': 'r2_test2'}, inplace=True)
df_additive3.rename(columns={'r2_test': 'r2_test3'}, inplace=True)
df_additive4.rename(columns={'r2_test': 'r2_test4'}, inplace=True)

df_additives = pd.concat((df_additive1[['molecule', 'num_layers', 'learning_rate', 'graph_readout', 'molecules_combination', 'training_loss', 'rmse_train1', 'rmse_test1', 'r2_train1', 'r2_test1']], df_additive2[['rmse_train2', 'rmse_test2', 'r2_train2', 'r2_test2']], df_additive3[['rmse_train3', 'rmse_test3', 'r2_train3', 'r2_test3']], df_additive4[['rmse_train4', 'rmse_test4', 'r2_train4', 'r2_test4']]), axis=1)

# -----

df_aryl1.rename(columns={'rmse_train': 'rmse_train1'}, inplace=True)
df_aryl2.rename(columns={'rmse_train': 'rmse_train2'}, inplace=True)
df_aryl3.rename(columns={'rmse_train': 'rmse_train3'}, inplace=True)

df_aryl1.rename(columns={'r2_train': 'r2_train1'}, inplace=True)
df_aryl2.rename(columns={'r2_train': 'r2_train2'}, inplace=True)
df_aryl3.rename(columns={'r2_train': 'r2_train3'}, inplace=True)

df_aryl1.rename(columns={'rmse_test': 'rmse_test1'}, inplace=True)
df_aryl2.rename(columns={'rmse_test': 'rmse_test2'}, inplace=True)
df_aryl3.rename(columns={'rmse_test': 'rmse_test3'}, inplace=True)

df_aryl1.rename(columns={'r2_test': 'r2_test1'}, inplace=True)
df_aryl2.rename(columns={'r2_test': 'r2_test2'}, inplace=True)
df_aryl3.rename(columns={'r2_test': 'r2_test3'}, inplace=True)

df_aryls = pd.concat((df_aryl1[['molecule', 'num_layers', 'learning_rate', 'graph_readout', 'molecules_combination', 'training_loss', 'rmse_train1', 'rmse_test1', 'r2_train1', 'r2_test1']], df_aryl2[['rmse_train2', 'rmse_test2', 'r2_train2', 'r2_test2']], df_aryl3[['rmse_train3', 'rmse_test3', 'r2_train3', 'r2_test3']]), axis=1)
```

```python
df_additives
```

```python
df_additives['mean_rmse_train'] = df_additives[['rmse_train1', 'rmse_train2', 'rmse_train3', 'rmse_train4']].mean(axis=1)
df_additives['std_rmse_train'] = df_additives[['rmse_train1', 'rmse_train2', 'rmse_train3', 'rmse_train4']].std(axis=1)

df_additives['mean_rmse_test'] = df_additives[['rmse_test1', 'rmse_test2', 'rmse_test3', 'rmse_test4']].mean(axis=1)
df_additives['std_rmse_test'] = df_additives[['rmse_test1', 'rmse_test2', 'rmse_test3', 'rmse_test4']].std(axis=1)

df_additives['mean_r2_train'] = df_additives[['r2_train1', 'r2_train2', 'r2_train3', 'r2_train4']].mean(axis=1)
df_additives['std_r2_train'] = df_additives[['r2_train1', 'r2_train2', 'r2_train3', 'r2_train4']].std(axis=1)

df_additives['mean_r2_test'] = df_additives[['r2_test1', 'r2_test2', 'r2_test3', 'r2_test4']].mean(axis=1)
df_additives['std_r2_test'] = df_additives[['r2_test1', 'r2_test2', 'r2_test3', 'r2_test4']].std(axis=1)

# ----

df_aryls['mean_rmse_train'] = df_aryls[['rmse_train1', 'rmse_train2', 'rmse_train3']].mean(axis=1)
df_aryls['std_rmse_train'] = df_aryls[['rmse_train1', 'rmse_train2', 'rmse_train3']].std(axis=1)

df_aryls['mean_rmse_test'] = df_aryls[['rmse_test1', 'rmse_test2', 'rmse_test3']].mean(axis=1)
df_aryls['std_rmse_test'] = df_aryls[['rmse_test1', 'rmse_test2', 'rmse_test3']].std(axis=1)

df_aryls['mean_r2_train'] = df_aryls[['r2_train1', 'r2_train2', 'r2_train3']].mean(axis=1)
df_aryls['std_r2_train'] = df_aryls[['r2_train1', 'r2_train2', 'r2_train3']].std(axis=1)

df_aryls['mean_r2_test'] = df_aryls[['r2_test1', 'r2_test2', 'r2_test3']].mean(axis=1)
df_aryls['std_r2_test'] = df_aryls[['r2_test1', 'r2_test2', 'r2_test3']].std(axis=1)
```

### Learning Rate

```python
df_additives[['mean_rmse_test','learning_rate']]
```

```python
df_additives_lr = df_additives[['mean_rmse_test','learning_rate']]
df_additives_001 = df_additives_lr[df_additives_lr['learning_rate'] == 0.01]
df_additives_0001 = df_additives_lr[df_additives_lr['learning_rate'] == 0.001]
```

```python
df_additives_lr
```

```python
plt.figure(figsize=(15,8))

x001 = list(range(1,len(df_additives_001['mean_rmse_test'].to_numpy())+1))
y001 = df_additives_001['mean_rmse_test'].to_numpy()

x0001 = list(range(1,len(df_additives_0001['mean_rmse_test'].to_numpy())+1))
y0001 = df_additives_0001['mean_rmse_test'].to_numpy()
plt.scatter(x001, y001, label='Learning Rate = 0.01',color='b')
plt.scatter(x0001, y0001, label='Learning Rate = 0.001',color='r')
plt.title('GNN Test')
plt.xlabel('Results (From Best to Worst)')
plt.ylabel('RMSE (%)')
plt.legend()
plt.show()
```

```python
list(range(1,len(df_additives_lr['mean_rmse_test'].to_numpy())+1))
```

```python
plt.figure(figsize=(15, 8))
plt.xticks(rotation=0)
sns.boxplot(x = df_additives_lr['learning_rate'], y = df_additives_lr['mean_rmse_test'])
sns.swarmplot(x = df_additives_lr['learning_rate'], y = df_additives_lr['mean_rmse_test'])
plt.title('GNN Test: Learning Rates')
plt.xlabel('Learning Rate')
plt.ylabel('RMSE (%)')
plt.show()
```

```python
plt.figure(figsize=(15,8))

lrs = df_additives_lr['learning_rate'].to_numpy()

x = list(range(1,len(df_additives_lr['mean_rmse_test'].to_numpy())+1))
  
rmse = df_additives_lr['mean_rmse_test'].to_numpy()
  
# empty list, will hold color value
col = []
  
for i in range(0, len(lrs)):
    if lrs[i]==0.01:
        col.append('blue')  
    else:
        col.append('red') 

plt.scatter(x, rmse, c = col, s = 20,
            linewidth = 3)
plt.title('GNN Test: Learning Rates')
plt.xlabel('Results (From Best to Worst)')
plt.ylabel('RMSE (%)')
plt.show()
```

### Number of Graph Convolutional Layers

```python
df_additives[['num_layers', 'mean_rmse_test']]
```

```python
df_additives_layers = df_additives[['num_layers', 'mean_rmse_test']]
```

```python
plt.figure(figsize=(15,8))

gcn = df_additives_layers['num_layers'].to_numpy()

x = list(range(1,len(df_additives_layers['mean_rmse_test'].to_numpy())+1))
  
rmse = df_additives_layers['mean_rmse_test'].to_numpy()
  
# empty list, will hold color value
col = []
  
for i in range(0, len(gcn)):
    if gcn[i]==2:
        col.append('blue')  
    elif gcn[i]==3:
        col.append('red') 
    elif gcn[i]==4:
        col.append('yellow') 
    elif gcn[i]==5:
        col.append('green') 

plt.scatter(x, rmse, c = col, s = 20,
            linewidth = 3)
plt.title('GNN Test: Number of Graph Convolutional Layers')
plt.xlabel('Results (From Best to Worst)')
plt.ylabel('RMSE (%)')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.xticks(rotation=0)
sns.boxplot(x = df_additives_layers['num_layers'], y = df_additives_layers['mean_rmse_test'])
sns.swarmplot(x = df_additives_layers['num_layers'], y = df_additives_layers['mean_rmse_test'])
plt.title('GNN Test: Number of Graph Convolutional Layers')
plt.xlabel('Number of Graph Convolutional Layers')
plt.ylabel('RMSE (%)')
plt.show()
```

### Graph Read-out

```python
df_additives[['graph_readout', 'mean_rmse_test']]
```

```python
df_additives_readout = df_additives[['graph_readout', 'mean_rmse_test']]
```

```python
plt.figure(figsize=(15,8))

readout = df_additives_readout['graph_readout'].to_numpy()

x = list(range(1,len(df_additives_readout['mean_rmse_test'].to_numpy())+1))
  
rmse = df_additives_readout['mean_rmse_test'].to_numpy()
  
# empty list, will hold color value
col = []
  
for i in range(0, len(readout)):
    if readout[i]=='sum':
        col.append('blue')  
    elif readout[i]=='mean':
        col.append('red') 
    elif readout[i]=='max':
        col.append('green') 

plt.scatter(x, rmse, c = col, s = 20,
            linewidth = 3)
plt.title('GNN Test: Graph Read-out')
plt.xlabel('Results (From Best to Worst)')
plt.ylabel('RMSE (%)')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.xticks(rotation=0)
sns.boxplot(x = df_additives_readout['graph_readout'], y = df_additives_readout['mean_rmse_test'])
sns.swarmplot(x = df_additives_readout['graph_readout'], y = df_additives_readout['mean_rmse_test'])
plt.title('GNN Test: Graph Read-out')
plt.xlabel('Graph Read-out Method')
plt.ylabel('RMSE (%)')
plt.show()
```
